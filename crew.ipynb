```python
from dotenv import load_dotenv
load_dotenv()
from crewai import Agent, Task , Crew
from crewai.tools import BaseTool
from crewai import LLM
import os
```

```python
llm=LLM(
    model="ollama/llama3.2:1b",
    api_base="http://localhost:11434",
    temperature=0.2,
    )

print(llm.call("hello"))
```

```python
research_agent=Agent(
    role="Reseach Agent",
    goal="Research interesting facts about the topic:{topic}",
    backstory="you are an expert at finding relevant and factual data",
    verbose=True,
    llm=llm
)

writer_agent=Agent(
    role="creative writer",
    goal="write a short blog summary using research",
    backstory="you are skilled at writing engaging summaries based on provided content",
    llm=llm,
    verbose=True,

)
```

```python
task1=Task(
    description="find 3-5 intersting and recent facts about {topic}",
    expected_output="a bullet of 3-5 facts",
    agent=research_agent,
  
)
print(task1)
task_2=Task(
    description="write a short blog summary about {topic} using the research provided",
    expected_output="a engaging blog summary",
    agent=writer_agent,
    context=[task1],
    
)
print(task_2)
```

```python
crew=Crew(
    agents=[research_agent,writer_agent],
    tasks=[task1,task_2],
    verbose=True,

)
print(crew)
```

```python
crew.kickoff(inputs={"topic":" future of AI"})
```
