```python
from dotenv import load_dotenv
load_dotenv()
from crewai import Agent, Task, Crew

from crewai import LLM 


```

```python
llm=LLM(
    model="ollama/llama3.2:1b",
    api_base="http://localhost:11434",
    temperature=0.2
)
llm.call("future of ev cars")
```

```python
from crewai import LLM
from crewai_tools import SerperDevTool
research_agent=Agent(
 role="Research Specialist",
  goal="Resarch intersecting facts about the topics:{topic}",
  backstory="ypu are an expert at finding relevant and factual data",
  tools=[SerperDevTool()],
  verbose=True,
  llm=llm
)

print(research_agent)
writer_agent=Agent(
    role="Creative writer",
    goal="write a short blog summary using the research",
    backstory="you are an skilled at writing engaging summaries based on provided content",
    llm=llm,
    verbose=True
)
```

```python
task1=Task(
description="find 3.5 intersting facts about the topics {topic}",
expected_output="a bullt list of 3-5 facts",
agent=research_agent,

)

task2=Task(
    description="write a 100 word blog post summary about {topic}",
    expected_output=" a blog post summary",
    context=[task1],
    agent=writer_agent,
)
```

```python
crew=Crew(
    agents=[research_agent,writer_agent],
    tasks=[task1,task2],
    verbose=True,
)
```

```python
crew.kickoff(inputs={"topic":"future of EV cars"})
```
