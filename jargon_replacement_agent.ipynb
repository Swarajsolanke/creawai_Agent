```python
from dotenv import load_dotenv
load_dotenv()
from crewai import Agent, Task , Crew
from crewai.tools import BaseTool
```

```python
class JargonReplcement(BaseTool):
    name:str="JargonReplcement"
    description:str="A tool to replace jargon words with simpler alternatives"

    def _run(self, email:str)->str:
        replacements={
            "PRX":"project phonix (internal revmaap project)",
            "TAX":"technical architecture stack",
            "DBX":"client database cluster",
            "SDS":"smart data syncre",
            "POC":"proof of concept",
            "SYNCBOT":"inter assistant bot",
             "ping":"reach out"
            }
        suggestion=[]
        email_lower=email.lower()
        for jargon, replacement in replacements.items():
            if jargon.lower() in email_lower:
                suggestion.append(f" consider replacing '{jargon}' with '{ replacement}'")
        return "\n".join(suggestion) if suggestion else "no jargons found"
    

jr=JargonReplcement()





```

```python
from crewai import LLM
llm=LLM(
    model="gemini/gemini-2.0-flash",
    temperature=0.1,
)
llm.call("how are you")

```

```python
email_assistant=Agent(
    role="Email Assistant Agent",
    goal="Improve emaile and make then sound professional and clear and also concise one ",
    backstory="A highly experienced communication exper skills in write professional email for the client",
    verbose=True,
    tools=[jr],
    llm=llm
)

original_email=" looping in priyas . TAS and PRX updates are in the deck. ETA for SDS integration is friday let's sync  up  tomorrow if syncbot allows. ping me if anu blockers."



jr.run(original_email)
```

```python
email_task=Task(
    description=f"""take the following email and rewrite it into professional and polish version"
'''{original_email}'''""",
agent=email_assistant,
expected_output="A professional, clear, and concise email")

crew=Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True
)

result=crew.kickoff()
print(result)


```
